## Stock Market Data Collection
The purpose of the python notebook is to collect historical stock market data for companies of different industry. THe historical data will then be stored as .csv files to be stored in 

The stocks that will be used are:
* UGI Corporation (UGI), utilities
* Amazon, (AMZN), tech
* Rio Tinto (RIO), materials
* General Electric (GE), industrials
* Pfizer (PFE), health care
* Goldman Sachs (GS), financials
* BP (BP), energy,
* Proctor & Gamble (PG), consumer staples
* Disney (DIS), consumer discretionary
* S&P500, Market Index

```python
# Import Statements for libraries
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Use yfinance library to load Tickers
ugi = yf.Ticker('UGI')
amzn = yf.Ticker('AMZN')
rio = yf.Ticker('RIO')
ge = yf.Ticker('GE')
pfe = yf.Ticker('PFE')
gs = yf.Ticker('GS')
bp = yf.Ticker('BP')
pg = yf.Ticker('PG')
dis = yf.Ticker('DIS')
sp500 = yf.Ticker('^GSPC')
```

At this point, I have loaded ticker information. Now, I want to select the historical data. I will focus on data from the last five years. This is achieved using the history function from the yfinance library.

```python
ugi_historical = ugi.history(period = '5y')
amzn_historical = amzn.history(period = '5y')
rio_historical = rio.history(period = '5y')
ge_historical = ge.history(period = '5y')
pfe_historical = pfe.history(period = '5y')
gs_historical = gs.history(period = '5y')
bp_historical = bp.history(period = '5y')
pg_historical = pg.history(period = '5y')
dis_historical = dis.history(period = '5y')
sp500_historical = sp500.history(period = '5y')
```

```python
# Modifying the data frame column titles

def TickerVector(len, ticker):
    vec = []
    for i in range(len):
        vec.append(ticker)
    return vec


ugi_historical['ticker'] = TickerVector(len(ugi_historical),'UGI')
amzn_historical['ticker'] = TickerVector(len(amzn_historical), 'AMZN')
rio_historical['ticker'] = TickerVector(len(rio_historical), 'RIO')
ge_historical['ticker'] = TickerVector(len(ge_historical), 'GE')
pfe_historical['ticker'] = TickerVector(len(pfe_historical), 'PFE')
gs_historical['ticker'] = TickerVector(len(gs_historical), 'GS')
bp_historical['ticker'] = TickerVector(len(bp_historical), 'BP')
pg_historical['ticker'] = TickerVector(len(pg_historical), 'PG')
dis_historical['ticker'] = TickerVector(len(dis_historical), 'DIS')
sp500_historical['ticker'] = TickerVector(len(sp500_historical), '^GSPC')

ugi_historical = ugi_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
ugi_historical = ugi_historical.rename_axis('date')

amzn_historical = amzn_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
amzn_historical = amzn_historical.rename_axis('date')

rio_historical = rio_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
rio_historical = rio_historical.rename_axis('date')

ge_historical = ge_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
ge_historical = ge_historical.rename_axis('date')

pfe_historical = pfe_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
pfe_historical = pfe_historical.rename_axis('date')

gs_historical = gs_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
gs_historical = gs_historical.rename_axis('date')

bp_historical = bp_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
bp_historical = bp_historical.rename_axis('date')

pg_historical = pg_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
pg_historical = pg_historical.rename_axis('date')

dis_historical = dis_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
dis_historical = dis_historical.rename_axis('date')

sp500_historical = sp500_historical.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume",
                                        "Dividends":"dividends","Stock Splits":"splits"})
sp500_historical = sp500_historical.rename_axis('date')
```

```python
ugi_historical.to_csv('data/ugi.csv', index=True)
amzn_historical.to_csv('data/amzn.csv', index=True)
rio_historical.to_csv('data/rio.csv', index=True)
ge_historical.to_csv('data/ge.csv', index=True)
pfe_historical.to_csv('data/pfe.csv', index=True)
gs_historical.to_csv('data/gs.csv', index=True)
bp_historical.to_csv('data/bp.csv', index=True)
pg_historical.to_csv('data/pg.csv', index=True)
dis_historical.to_csv('data/dis.csv', index=True)
sp500_historical.to_csv('data/sp500.csv', index=True)
```
